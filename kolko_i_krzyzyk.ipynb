```python
from IPython.display import clear_output

runda = 0
ruch_wykozystany = '-'
wybor = ['','','','','','','','','']
ruch_mozliwy = [1,2,3,4,5,6,7,8,9]

print("Legenda:\n")
print("%3s  | %3s  | %3s" % ('1','2','3'))
print('-'*20)
print("%3s  | %3s  | %3s" % ('4','5','6'))
print('-'*20)
print("%3s  | %3s  | %3s\n" % ('7','8','9'))

while runda != 9:
    if runda % 2 == 0:
        znak = 'X'
        print("Do wyboru pozostaly pola: ", ruch_mozliwy)
        try:
            user1 = input("Podaj w którym miejscu ma zostać wstawiony znak X: ")
            user1 = int(user1)
        except:
            print("Podana wartość nie znajduje się na liście dostępnych. Podaj inną")
        while True:
            if user1 in ruch_mozliwy:
                ruch_mozliwy[user1 -1] = ruch_wykozystany
                wybor[user1 -1] = znak
                break
            else:
                try:
                    user1 = input("Podana wartosc został juz uzyta. Podaj inna ")
                    user1 = int(user1)
                except:
                    print("Podana wartość nie znajduje się na liście dostępnych. Podaj inną")
        if (wybor[0] + wybor[1] + wybor[2] == 'XXX' 
            or wybor[3] + wybor[4] + wybor[5] == 'XXX'
            or wybor[6] + wybor[7] + wybor[8] == 'XXX'
            or wybor[0] + wybor[3] + wybor[6] == 'XXX'
            or wybor[1] + wybor[4] + wybor[7] == 'XXX'
            or wybor[2] + wybor[5] + wybor[8] == 'XXX'
            or wybor[0] + wybor[4] + wybor[8] == 'XXX'
            or wybor[2] + wybor[4] + wybor[6] == 'XXX'):
            print('X win!')
            break
    else:
        znak = 'O'
        print("Do wyboru pozostaly pola: ", ruch_mozliwy)
        try:
            user2 = input("Podaj w którym miejscu ma zostać wstawiony znak O: ")
            user2 = int(user2)
        except:
            print("Podana wartość nie znajduje się na liście dostępnych. Podaj inną")
        while True:
            if user2 in ruch_mozliwy:
                ruch_mozliwy[user2 -1] = ruch_wykozystany
                wybor[user2 -1] = znak
                break
            else:
                try:
                    user2 = input("Podana wartosc został juz uzyta. Podaj inna ")
                    user2 = int(user2)
                except:
                    print("Podana wartość nie znajduje się na liście dostępnych. Podaj inną")
        if (wybor[0] + wybor[1] + wybor[2] == 'OOO' 
            or wybor[3] + wybor[4] + wybor[5] == 'OOO'
            or wybor[6] + wybor[7] + wybor[8] == 'OOO'
            or wybor[0] + wybor[3] + wybor[6] == 'OOO'
            or wybor[1] + wybor[4] + wybor[7] == 'OOO'
            or wybor[2] + wybor[5] + wybor[8] == 'OOO'
            or wybor[0] + wybor[4] + wybor[8] == 'OOO'
            or wybor[2] + wybor[4] + wybor[6] == 'OOO'):
            print('O win!')
            break
            
    print("%3s  | %3s  | %3s" % (wybor[0],wybor[1],wybor[2]))
    print('-'*20)
    print("%3s  | %3s  | %3s" % (wybor[3],wybor[4],wybor[5]))
    print('-'*20)
    print("%3s  | %3s  | %3s\n" % (wybor[6],wybor[7],wybor[8]))
    runda  += 1
#    clear_output(wait=True)
    
print("%3s  | %3s  | %3s" % (wybor[0],wybor[1],wybor[2]))
print('-'*20)
print("%3s  | %3s  | %3s" % (wybor[3],wybor[4],wybor[5]))
print('-'*20)
print("%3s  | %3s  | %3s" % (wybor[6],wybor[7],wybor[8]))
```
